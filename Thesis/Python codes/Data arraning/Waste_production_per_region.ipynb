```python
import pandas as pd
from IPython.display import display
import os

regions = ["Abruzzo","Basilicata","Calabria","Campania","EmiliaRomagna","FriuliVeneziaGiulia","Lazio","Liguria","Lombardia","Marche","Molise","Piemonte","Puglia","Sardegna","Sicilia","Toscana","TrentinoAltoAdige","Umbria","ValledAosta","Veneto"]

industria_alimentare = []
totale_industria = []
rifiuti_alimentari = []
totale_rifiuti = []

for region in regions:
    df = pd.read_csv(f"/content/drive/MyDrive/Thesis/Python codes/Waste Production per Region/{region}/table_1_{region}.csv")
    index_01 = df.index[df[df.columns[0]].str.contains("10, 11 - Industria alimentare e delle bevande")] # index_01 is a Pandas Index object that contains the index labels of rows where the first column contains the specified string
    index_02 = df.index[df[df.columns[0]].str.contains("Totale")]
    industria_alimentare.append(df.iloc[index_01[0]][3]) # df.iloc[index_01[0]] returns the list of objects present in the corresponding rows, df.iloc[index_01[0]][3] selects the value that I need, in this case the one in the 4th column
    totale_industria.append(df.iloc[index_02[0]][3])
    
    
    df = pd.read_csv(f"/content/drive/MyDrive/Thesis/Python codes/Waste Production per Region/{region}/table_2_{region}.csv")
    index_01 = df.index[df[df.columns[1]].str.contains("Rifiuti prodotti da agricoltura, orticoltura, acquacoltura, selvicoltura, caccia e pesca, trattamento e preparazione di alimenti")]
    index_02 = df.index[df[df.columns[0]].str.contains("Totale")]
    rifiuti_alimentari.append(df.iloc[index_01[0]][4])
    totale_rifiuti.append(df.iloc[index_02[0]][4])

    
df = pd.DataFrame()
df.insert(0, 'NOME_REG', regions)
df.insert(1, 'ATECO 10,11 [t]', industria_alimentare)
df.insert(2, 'Totale industria', totale_industria)
df['ATECO 10,11 [t]']= [s.replace(".", "") for s in df['ATECO 10,11 [t]'].astype(str)]
df['Totale industria']= [s.replace(".", "") for s in df['Totale industria'].astype(str)]
df['Peso industria'] = round(df['ATECO 10,11 [t]'].astype(int)/df['Totale industria'].astype(int)*100,2)


df1 = pd.DataFrame()
df1.insert(0, 'NOME_REG', regions)
df1.insert(1, 'EER 02 [t]', rifiuti_alimentari)
df1.insert(2, 'Totale rifiuti', totale_rifiuti)
df1['EER 02 [t]']= [s.replace(".", "") for s in df1['EER 02 [t]'].astype(str)]
df1['Totale rifiuti']= [s.replace(".", "") for s in df1['Totale rifiuti'].astype(str)]
df1['Peso rifiuti'] = round(df1['EER 02 [t]'].astype(int)/df1['Totale rifiuti'].astype(int)*100,2)

display(df,df1)

path_data_arranged = '/content/drive/MyDrive/Thesis/Python codes/Waste Production per Region'
df.to_csv(os.path.join(path_data_arranged, 'Food Waste per ATECO.csv'))
df1.to_csv(os.path.join(path_data_arranged, 'Food Waste per EER.csv'))


```

```python
import pandas as pd
from IPython.display import display

regions = ["Abruzzo","Basilicata","Calabria","Campania","EmiliaRomagna","FriuliVeneziaGiulia","Lazio","Liguria","Lombardia","Marche","Molise","Piemonte","Puglia","Sardegna","Sicilia","Toscana","TrentinoAltoAdige","Umbria","ValledAosta","Veneto"]
counter = 0
for region in regions:
  for i in range(1,3):
    print(counter)
    # Load the table into a dataframe
    df1 = pd.read_csv(f"/content/drive/MyDrive/Thesis/Python codes/Waste Production per Region/{region}/table_{i}_{region}.csv")

    # Get the values in the first column and extract the numbers
    col1 = df1.iloc[:, 0].values
    numbers = []
    for val in col1:
        if "-" in val:
            split_val = val.split("-")[0]
            if "," in split_val:
                split_val = split_val.split(",")[0]
        else:
          if "Totale" in val:
            continue
          split_val = val 
        if "ND" in split_val:
                continue       
        try:
            numbers.append(split_val.zfill(2))
        except ValueError:
            pass
    numbers = [s.replace(" ", "") for s in numbers] # remove spaces from the numbers

    numbers_temp = []
    for number in numbers:
      numbers_temp.append(int(number))

    # Find the missing values in the progression
    missing = []
    for i in range(1, max(numbers_temp) + 1):
        if str(i).zfill(2) not in numbers:
            missing.append(str(i).zfill(2))

    # Create a dataframe with the missing values and default values for the other columns
    missing_df = pd.DataFrame({df1.columns[0]: [str(i) + " missing" for i in missing],
                              df1.columns[1]: [None for i in range(len(missing))]})

    # Concatenate the original dataframe with the missing values dataframe
    df = pd.concat([df1, missing_df], sort=False)

    # Sort the dataframe based on the values in the first column
    df.sort_values(by=df.columns[0], inplace=True)

    # Reset the index of the dataframe
    df.reset_index(drop=True, inplace=True)

    # Save the dataframe to a csv file
    #df.to_csv("table_filled.csv", index=False)
    counter += 1
    display(df) 
```
