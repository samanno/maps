```python
import pandas as pd
import numpy as np

# Read the first CSV file into a dataframe
df1 = pd.read_csv('/content/drive/MyDrive/Thesis/Python codes/Food Industry Enterprises per Province/Number of Employees.csv')

# Read the second CSV file into a dataframe
df2 = pd.read_csv('/content/drive/MyDrive/Thesis/Python codes/Food Industry Enterprises per Province/Number of Enterprises.csv')

def is_number(x):
    """
    It checks if the element of a dataframe is a number and returns a boolean
    """
    try:
        float(x)
        return True
    except:
        return False

# Function to divide two dataframes
def divide_dataframe(df1, df2):
    """
    It checks first if the [i,j] position of both dataframes is not a NaN; then, when this first condition is satisfied, it checks if the [i,j] position of both dataframes
    is a number (by recalling the function "is_number"). If also this second condition is satisfied, it performs the operation. While, if the "is_number" condition is not satisfied,
    it assigns to the result the value present in the [i,j] position of the first dataframe. The "elif" and "else" related to the first conditional are the situations in which the
    "pd.notna" verifies for only one of the two input dataframes.
    """
    df3 = df1.copy()
    for i in range(df1.shape[0]): # .shape returns the dimensions of the object [0] = rows, [1] = columns
        for j in range(df1.shape[1]):
            if pd.notna(df1.iloc[i, j]) and pd.notna(df2.iloc[i, j]):
                if is_number(df1.iloc[i, j]) and is_number(df2.iloc[i, j]) and float(df2.iloc[i, j]): # float is to ensure to avoid a by 0 division
                    df3.iloc[i, j] = round((float(df1.iloc[i, j]) / float(df2.iloc[i, j])),2)
                else:
                  df3.iloc[i, j] = df1.iloc[i, j]
                    
            elif pd.notna(df1.iloc[i, j]):
                df3.iloc[i, j] = df1.iloc[i, j]
            else:
                df3.iloc[i, j] = df2.iloc[i, j]
    df3 = df3.applymap(lambda x: x if isinstance(x, str) else float(x) if is_number(x) else np.nan) # it's usefull in case there are missing values or divisions by 0 that can generate a NaN
    return df3

# Call the function and print the result
df3 = divide_dataframe(df1, df2)

df3.to_csv('/content/drive/MyDrive/Thesis/Python codes/Food Industry Enterprises per Province/Employees per enterprise.csv', index = False)
display(df3)


```
