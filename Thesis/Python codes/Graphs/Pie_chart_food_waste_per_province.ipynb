#For each province and the whole region


```python
import pandas as pd
import matplotlib.pyplot as plt

# Read in the data from the CSV file
csv_file = '/content/drive/MyDrive/Thesis/Python codes/Waste production per province - ER/Food Waste per ATECO.csv'
df = pd.read_csv(csv_file)

# Select the relevant columns from the DataFrame
df = df.iloc[:, 2:]

# Define the list of provinces to plot
provinces = ['Bologna', 'Ferrara', 'Forlì Cesena', 'Modena', 'Parma', 'Piacenza', 'Ravenna', 'Reggio Emilia', 'Rimini', 'Emilia Romagna']

# Loop through each row of the DataFrame and create a pie chart
for index, row in df.iterrows():
    # Calculate the contribution of each sector to the total waste
    total = row[-1]
    non_zero_contributions = row[:-1][row[:-1] != 0]  # filter out zero contributions
    non_zero_labels = non_zero_contributions.index

    explode = [0.1] * len(non_zero_contributions)
    
    # Set the sizes for the pie chart
    sizes = non_zero_contributions / total
    
    # Create the pie chart with the appropriate labels and title
    fig = plt.figure(figsize=(15, 15))  # increase the size of the chart
    plt.pie(sizes, labels=non_zero_labels, autopct='%1.1f%%', explode=explode, 
              pctdistance=0.90, labeldistance=1.05, wedgeprops=dict(width=0.6, edgecolor='w'), textprops={'fontsize': 12})
    plt.legend(non_zero_labels, loc="best")
    plt.title(provinces[index])
    plt.show()

    path = f"/content/drive/MyDrive/Thesis/Graphs/Waste production per province/Pie Charts/Food waste per ATECO_{provinces[index]}"
    fig.savefig(path, dpi=300, bbox_inches='tight')

# Display the DataFrame
display(df)
```

#For each ATECO or EER category

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read in the data from the CSV file
csv_file = '/content/drive/MyDrive/Thesis/Python codes/Waste production per province - ER/Food Waste per ATECO.csv'
df = pd.read_csv(csv_file)

# Select the relevant columns from the DataFrame
df = df.iloc[:, 2:]

# Define the list of provinces to plot
provinces = ['Bologna', 'Ferrara', 'Forlì Cesena', 'Modena', 'Parma', 'Piacenza', 'Ravenna', 'Reggio Emilia', 'Rimini']

for col in df.columns[:]:
    # Calculate the contribution of each sector to the total waste
    total = df[col].iloc[-1]
    non_zero_contributions = df[col].iloc[:-1][df[col].iloc[:-1] != 0]  # filter out zero contributions
    non_zero_labels = [provinces[i] for i in non_zero_contributions.index]

    explode = [0.1] * len(non_zero_contributions)

    # Set the sizes for the pie chart
    sizes = non_zero_contributions / total

    # Create the pie chart with the appropriate labels and title
    fig = plt.figure(figsize=(15, 15))  # increase the size of the chart
    plt.pie(sizes, labels=non_zero_labels, autopct='%1.1f%%', explode=explode, 
              pctdistance=0.90, labeldistance=1.05, wedgeprops=dict(width=0.6, edgecolor='w'), textprops={'fontsize': 12})
    plt.legend(non_zero_labels, loc="best")
    plt.title(f'ATECO CODE {col}')
    plt.show()

    path = f"/content/drive/MyDrive/Thesis/Graphs/Waste production per province/Pie Charts/ATECO CODE_{col}.png"
    fig.savefig(path, dpi=300, bbox_inches='tight')

display(df)
```
