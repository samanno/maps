```python
def extract_word_from_url(url):
  match = re.search("imp=(\w+)", url)
  if match:
    return match.group(1)
  else:
    return None
```

```python
def modify_url(url, word, number):
  modified_url = re.sub("impid=\d+", "impid=" + str(number), url)
  modified_url = re.sub("imp=\w+", "imp=" + word, modified_url)
  return modified_url
```

```python
url1 = "https://www.catasto-rifiuti.isprambiente.it/index.php?pg=findrsproduzione&width=1920&height=1080"
tablex = pd.read_html(url1, flavor="html5lib")
regions = []
for i in range(1, 21):
    regions.append(tablex[i][1].tolist()[0].replace(" ", "").replace("â€™", ""))
print(regions)
```

```python
url = "https://www.catasto-rifiuti.isprambiente.it/index.php?pg=rsproduzionedettaglio&aa=2019&regid=1&impid=01&imp=Piemonte"
counter = 1
for region in regions:
    try:
        os.makedirs(region)
    except FileExistsError:
        print(f"Directory {region} already exists.")
    url = modify_url(url, region, counter)
    tables = pd.read_html(url, flavor="html5lib")
    counter += 1
    for i, table in enumerate(tables):
        df = pd.DataFrame(table)
        df.to_csv(os.path.join(region,f'table_{i}_{extract_word_from_url(url)}.csv'), index=False)
```
